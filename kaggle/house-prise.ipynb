```python
hpc = hp.corr(numeric_only=True)
hpc
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(30,30))
sns.heatmap(hpc, annot=True)
```

```python
hps = hp.corr(numeric_only=True)
f, ax = plt.subplots(figsize=(30, 30))
sns.heatmap(hps, vmin=.5,vmax=1, square=True, annot=True);
```
